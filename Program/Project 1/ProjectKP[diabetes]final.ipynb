This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective is to predict based on diagnostic measurements whether a patient has diabetes. https://www.kaggle.com/datasets/mathchi/diabetes-data-set

Detail desc:

Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.
* Pregnancies: Number of times pregnant
* Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
* BloodPressure: Diastolic blood pressure (mm Hg)
* SkinThickness: Triceps skin fold thickness (mm)
* Insulin: 2-Hour serum insulin (mu U/ml)
BMI: Body mass index (weight in kg/(height in m)^2)
* DiabetesPedigreeFunction: Diabetes pedigree function
* Age: Age (years)
* Outcome: Class variable (0 or 1)

## Import Library dan Dataset

```python
# standard imports

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import sklearn

# import dataset
diabetes = pd.read_csv('diabetes.csv')

# view data
diabetes.head()
```

## Missing Value

```python
# mencari missing value
diabetes.isna()
```

```python
# menghitung missing value
diabetes.isna().sum()
```

## Pemisahan Data

* memisahkan data kolom fitur (X) dan data outcome (y)
* memisahkan data training dan data testing

```python
# memisahkan data menjadi fitur (X) dan label (y)

# X (kolom fitur tanpa target)
X = diabetes.drop('Outcome', axis=1)

# y (kolom target)
y = diabetes['Outcome']

X
```

```python
y
```

```python
# memisahkan data: train set dan test set
from sklearn.model_selection import train_test_split

# train_test_split menggunakan 25% dari data untuk test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

# bentuk data (baris, kolom)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Klasifikasi dan Prediksi

Metode Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X=X_train, y=y_train)
```

```python
# membuat prediksi
y_preds = clf.predict(X=X_test)
```

## Evaluasi

Menghitung akurasi

```python
# evaluasi model pada training set
train_acc = clf.score(X=X_train, y=y_train)
print(f"Akurasi model pada training dataset adalah: {train_acc*100}%")
```

```python
# evaluasi model pada test set
test_acc = clf.score(X=X_test, y=y_test)
print(f"Akurasi model pada testing dataset adalah: {test_acc*100:.2f}%")
```

## Correlation Matrix

Menunjukkan korelasi (hubungan) antara satu fitur dengan fitur lainnya.


```python
# Correlation Matrix
diabetes.corr()
```

Bernilai antara -1 dan 1:
* semakin mendekati +1 menunjukkan korelasi positif (jika fitur 1  meningkat, maka fitur 2 juga meningkat).
* semakin mendekati -1 menunjukkan korelasi negatif (jika fitur 1 meningkat, fitur 2 menurun).
* 0 menunjukkan tidak ada korelasi (kedua fitur tidak saling berhubungan).

```python
import seaborn as sns

corr_matrix = diabetes.corr()
fig, ax = plt.subplots(figsize=(15, 10))
plt.title('Correlation Matrix')
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="PuRd");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

## Confusion Matrix

Bagian ini merupakan visualisasi dari hasil prediksi yang dilakukan pada dataset menggunakan metode Random Forest.

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# confusion matrix (untuk visualisasi hasil prediksi)
cm = confusion_matrix(y_test, y_preds)

# visualisasi heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='PiYG',
            xticklabels=['Tidak Diabetes', 'Diabetes'],
            yticklabels=['Tidak Diabetes', 'Diabetes'])
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.title('Confusion Matrix')
plt.show()
```

## Visualisasi Akurasi

Hasil akurasi divisualisasikan menggunakan grafik, diambil dari library matplotlib.

```python
import matplotlib.pyplot as plt

# data untuk visualisasi
acc_scores = [train_acc, test_acc]
labels = ['Akurasi Data Training', 'Akurasi Data Test']

# membuat grafik visualisasi
plt.figure(figsize=(8, 5))
plt.bar(labels, acc_scores, color=['palevioletred', 'pink'])
plt.ylim(0, 1)  # set limit untuk sumbu y dari 0 hingga 1
plt.title('Akurasi Model Training dan Test Set')
plt.ylabel('Akurasi')
plt.show()
```
