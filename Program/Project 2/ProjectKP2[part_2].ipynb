## Frekuensi Penyakit Jantung Berdasarkan Jenis Nyeri Dada (cp)

```python
pd.crosstab(df.cp, df.target)
```

```python
# visualisasi
# frekuensi penyakit jantung per jenis nyeri dada
pd.crosstab(df.cp, df.target).plot(kind="bar",
                                   figsize=(10, 6),
                                   color=["salmon", "lightblue"])

plt.title("Frekuensi Penyakit Jantung Per Jenis Nyeri Dada")
plt.xlabel("Jenis Nyeri Dada")
plt.ylabel("Jumlah")
plt.legend(["Tidak Ada Penyakit", "Ada Penyakit"])
plt.xticks(rotation=0);
```

Grafik ini menampilkan frekuensi penyakit jantung berdasarkan jenis nyeri dada (chest pain - cp) dalam kategori:
* 0:  *Typical angina*: nyeri dada yang berhubungan dengan penurunan suplai darah ke jantung
* 1: *Atypical angina*: nyeri dada yang tidak berhubungan dengan jantung
* 2: *Non-anginal pain*: tipikal kejang esofagus (tidak berhubungan dengan jantung)
* 3: *Asymptomatic*: nyeri dada tanpa gejala, tidak menunjukkan tanda-tanda penyakit

untuk melihat hubungan antara gejala (nyeri dada) dan penyakit jantung. hasilnya menunjukkan bahwa pada jenis nyeri dada 0 yang dianggap sebagai tanda masalah jantung, mayoritas pasien justru lebih banyak yang tidak mengalami penyakit jantung.

## PEMODELAN


```python
df.head()
```

```python
# pemisahan data, jadi X dan y

X = df.drop("target", axis=1)
y = df["target"]
```

```python
X #data X mencakup fitur tanpa target
```

```python
y #data y itu data target
```

```python
np.random.seed(42)

# pemisahan data, jadi data train set dan test set
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
```

```python
X_train
```

```python
y_train, len(y_train)
```

```python
# memasukkan model
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}

# fungsi menyesuaikan dan menilai model
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """

    np.random.seed(42)
    model_scores = {}

    for name, model in models.items():
        model.fit(X_train, y_train) # penyesuaian model dan data
        model_scores[name] = model.score(X_test, y_test) # evaluasi model
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

## Perbandingan Model

```python
model_compare = pd.DataFrame(model_scores, index=["accuracy"])
model_compare.T.plot.bar();
```
