## Frekuensi Penyakit Jantung Berdasarkan Jenis Kelamin (sex)

```python
df.sex.value_counts()
```

```python
pd.crosstab(df.target, df.sex)
```

```python
# visualisasi
# frekuensi penyakit jantung berdasarkan jenis kelamin (sex)

pd.crosstab(df.target, df.sex).plot(kind="bar",
                                    figsize=(10, 6),
                                    color=["salmon", "lightblue"])

plt.title("Frekuensi Penyakit Jantung Berdasarkan Jenis Kelamin")
plt.xlabel("0 = Tidak ada penyakit, 1 = Ada penyakit")
plt.ylabel("Jumlah")
plt.legend(["Perempuan", "Laki-laki"]);
plt.xticks(rotation=0);
```

## Usia vs. Denyut Jantung Maks untuk Penyakit Jantung

```python
# visualiasi 2
# penyakit jantung dalam fungsi usia dan denyut jantung maksimal

plt.figure(figsize=(10, 6))

# penyebaran data positive penyakit jantung
plt.scatter(df.age[df.target==1],
            df.thalach[df.target==1],
            c="salmon")

# penyebaran data negative penyakit jantung
plt.scatter(df.age[df.target==0],
            df.thalach[df.target==0],
            c="lightblue")

plt.title("Penyakit Jantung dalam Fungsi Usia dan Denyut Jantung Maksimal")
plt.xlabel("Usia")
plt.ylabel("Denyut Jantung Maksimal")
plt.legend(["Ada Penyakit", "Tidak Ada Penyakit"]);
```
